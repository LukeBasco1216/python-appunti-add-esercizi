# ***b. Visualizzare il nome della provincia e il nome del comune***

```python
province = geopd.read_file("/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Limiti01012021_g/ProvCM01012021_g/ProvCM01012021_g_WGS84.dbf")
comuni = geopd.read_file("/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Limiti01012021_g/Com01012021_g/Com01012021_g_WGS84.dbf")
```

# ***d. Visualizzare numero di comuni per ogni provincia***


```python
df_prov_comune3 = pd.merge(province, comuni, how ="outer", on="COD_PROV")
df_prov_comune3
```

```python
num_comuni_in_province = df_prov_comune3.groupby("DEN_UTS", as_index = False)["COMUNE"].count()
num_comuni_in_province
```

# ***e. Visualizzare numero di comuni per ogni regione***

```python
df_reg_com = pd.merge(regioni, comuni, how = "outer", on="COD_REG")
df_reg_com
```

```python
num_comuni_in_reg = df_reg_com.groupby("DEN_REG", as_index = False)["COMUNE"].count()
num_comuni_in_reg
```

# ***f. Visualizzare numero di province per ogni regione***
